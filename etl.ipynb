```python
#quitar las columnas que no se van a usar
report.drop(columns=['Unnamed: 0', 'flight_no', 'registration', 'cn_ln', 'ground', 'summary', 
                            'ac_type', 'route', 'OperadOR', 'Ruta','HORA declarada'], inplace=True)
```

```python
#obtiene el anio de cada accidente
report["year"] = pd.to_datetime(report["fecha"]).dt.year

#selecciona los accidentes que pasaron desde el 2000 en adelante
report = report[report["year"] >= 2000]
```

```python
report = report.drop("fecha", axis=1)
```

```python
report.isnull().sum()
```

```python
report = report.replace(to_replace="?", value=np.nan)
```

```python
report.reset_index(drop=True, inplace=True)
```

```python
report = report.rename(columns={'all_aboard': 'total_abordo', 'PASAJEROS A BORDO': 'pasajeros_abordo',
                                'crew_aboard' : 'personal_abordo','cantidad de fallecidos':'total_fallecidos',
                                'passenger_fatalities' :'pasajeros_fallecidos', 'crew_fatalities':'personal_fallecido','year':'anio'})
```

```python
report
```

```python
report1 = report[(report["anio"] >= 2000) & (report["anio"] < 2011)]
report2 = report[(report["anio"] >= 2011) & (report["anio"] <= 2021)]
```

```python
report.to_csv("Accidentes_2000_al_actual.csv")
report1.to_csv("Accidentes_2000-2010.csv")
report2.to_csv("Accidentes_2011-2021.csv")
```
